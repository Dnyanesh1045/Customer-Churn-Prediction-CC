# creating function to generate IQR, lower limit, and Upper limit

```python
# Creating Function to generate IQR, lower limit and Upper limit
def find_outlier(df,feature):
    print("Outlier"+ feature)
    q1 = df[feature].quantile(0.25)
    q2 = df[feature].quantile(0.50)
    q3 = df[feature].quantile(0.75)
    iqr = q3-q1
    limit = iqr*1.5
    print(f"IQR:{iqr}")
    
    Lower_limit=q1-limit
    if Lower_limit < 0:
        Lower_limit = 0
    upper_limit = q3+limit
    print(f"Lower_limit:{Lower_limit}")
    print(f"median:{q2}")
    print(f"Upper_Limit:{upper_limit}")
    print('______________________________________________')
        
    
```

# check IQR, upper limit, and lower limit for each feature

```python
# check IQR, upper limit, and lower limit for each feature

for i in num_feat :
    find_outlier(num_feat, i)
```

# Modeling and Evaluation

```python
# Import Library for Modeling and Evaluations
from sklearn.model_selection import train_test_split,cross_val_score, RandomizedSearchCV
from sklearn.model_selection import GridSearchCV, KFold,StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier,GradientBoostingClassifier

import category_encoders as ce
from sklearn.preprocessing import StandardScaler,RobustScaler
from sklearn.compose import  ColumnTransformer
from sklearn.pipeline import Pipeline
from category_encoders import BinaryEncoder
from sklearn.metrics import f1_score,roc_auc_score,roc_curve,accuracy_score,precision_score,recall_score
```

```python
# Ordinal Variable
Income_Category_map = {
    'Less than $40K' : 0,
    '$40K - $60K'    : 1,
    '$60K - $80K'    : 2,
    '$80K - $120K'   : 3,
    '$120K +'        : 4,
    'Unknown'        : 5
}

Card_Category_map = {
    'Blue'     : 0,
    'Silver'   : 1,
    'Gold'     : 2,
    'Platinum' : 3
}

Education_Level_map = {
    'Uneducated'    : 0,
    'High School'   : 1,
    'College'       : 2,
    'Graduate'      : 3,
    'Post-Graduate' : 4,
    'Doctorate'     : 5,
    'Unknown'       : 6
}

df.loc[:, 'Income_Category'] = df['Income_Category'].map(Income_Category_map)
df.loc[:, 'Card_Category'] = df['Card_Category'].map(Card_Category_map)
df.loc[:, 'Education_Level'] = df['Education_Level'].map(Education_Level_map)

# Nominal Variable

Gender_map = {
    'M' : 0,
    'F' : 1
}

Attrition_Flag_map = {
    'Existing Customer' : 0,
    'Attrited Customer' : 1
}

Marital_Status_map = {
    'Single' : 0,
    'Married' : 1,
    'Divorced' : 2,
    'Unknown' : 3
}


df.loc[:, 'Gender'] = df['Gender'].map(Gender_map)
df['Dependent_count'].astype(int)
df.loc[:, 'Attrition_Flag'] = df['Attrition_Flag'].map(Attrition_Flag_map)
df.loc[:, 'Marital_Status'] = df['Marital_Status'].map(Marital_Status_map)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=2023)
X_train=X_train.reset_index(drop=True)
X_test=X_test.reset_index(drop=True)
```

```python
st_scaler=StandardScaler()
rb_scaler=RobustScaler()

transformer = ColumnTransformer([
    ("standard scaler",st_scaler,['Customer_Age', 'Months_on_book', 'Total_Revolving_Bal', 
                                             'Total_Amt_Chng_Q4_Q1', 'Total_Ct_Chng_Q4_Q1', 'Total_Trans_Ct']),
    ("robust scaler",rb_scaler,['Credit_Limit', 'Avg_Open_To_Buy', 
                                             'Total_Trans_Amt', 'Avg_Utilization_Ratio'])
],remainder = "passthrough")
```

```python
# We use robust scaler cause dataset have many outlier
```

# Modeling

```python
logreg = LogisticRegression(random_state=2023)
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier(random_state=2023)
rf = RandomForestClassifier(random_state=2023)
ab = AdaBoostClassifier(random_state=2023)
gb = GradientBoostingClassifier(random_state=2023)
```

```python
list_model = {'Logistic Regression' : logreg, 'KNN' : knn, 'Decision Tree' : dt, 'Random Forest' : rf, 
              'AdaBoost' :ab, 'GradientBoost' : gb}
```

```python
model_list = [logreg,knn,dt,rf,ab,gb]
```

```python
f1_score = []
mean_f1_score = []
std_f1_score = []
crossval = StratifiedKFold(n_splits=5)

def model_eval_f1_score(model):  # Pass Dictionary of models
    for i in model:
        estimator = Pipeline([
            ('tranformer',transformer),
            ('model',list_model[i])
        ])
        model_cv=cross_val_score(estimator,X_train,y_train,cv=crossval,scoring='f1',error_score='raise')
        f1_score.append(model_cv)
        mean_f1_score.append(model_cv.mean())
        std_f1_score.append(model_cv.std())
```

```python
roc_auc = []
mean_roc_auc = []
std_roc_auc = []
crossval = StratifiedKFold(n_splits=5)

def model_eval_roc_auc(model): # Pass Dictionary of models
    for i in model:
        estimator = Pipeline([
            ('tranformer',transformer),
            ('model',list_model[i])
        ])
        model_cv=cross_val_score(estimator,X_train,y_train,cv=crossval,scoring='roc_auc',error_score='raise')
        roc_auc.append(model_cv)
        mean_roc_auc.append(model_cv.mean())
        std_roc_auc.append(model_cv.std())
```

```python
model_eval_f1_score(list_model)  # # Pass Dictionary of models
```

```python
model_eval_roc_auc(list_model)    # Pass Dictionary of models
```

```python
model_b_tuning = pd.DataFrame({'model':['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest', 'AdaBoost', 'GradientBoost'],
    'f1_Score' : f1_score,
    'f1_score_Mean':mean_f1_score,
    'f1_score_std':std_f1_score,
    'roc_auc' : roc_auc,
    'roc_auc_Mean': mean_roc_auc,
    'roc_auc_Std': std_roc_auc
})

model_b_tuning
```

```python
recall = []
recall_mean = []
std_recall = []
crossval = StratifiedKFold(n_splits=5)

def model_eval_recall(model): # Pass Dictionary of models
    for i in model:
        estimator = Pipeline([
            ('tranformer',transformer),
            ('model',list_model[i])
        ])
        model_cv=cross_val_score(estimator,X_train,y_train,cv=crossval,scoring='recall',error_score='raise')
        recall.append(model_cv)
        recall_mean.append(model_cv.mean())
        std_recall.append(model_cv.std())
```

```python
recall_df = pd.DataFrame({'model':['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest', 'AdaBoost', 'GradientBoost'],
    'recall' : f1_score,
    'recall_mean':mean_f1_score,
    'std_recall':std_f1_score,
})

recall_df
```

## Hyperparameter Tuning


* Choose following method for hyperparameter tuning
    1. **RandomizedSearchCV** --> Fast
    2. **GridSearchCV**
* Assign hyperparameters in form of dictionery
* Fit the model
* Check best paramters and best score

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Randomized search CV

# Number of trees in Random Forest
n_estimators = [int(x) for x in np.linspace(start=100,stop=1200,num=12)]

# Number of feature to consider at every split
max_features=['auto','sqrt']

# Maximum depth of trees
max_depth= [int(x) for x in np.linspace(5,30,num=6)]

# Minimum number of sample required to split a node
min_samples_split = [2,5,10,15,100]

# Minimum number of sample required at each node
min_sample_leaf = [1,2,3,4,10]
```

```python
# Create the random grid

random_grid={"n_estimators":n_estimators,
             "max_features":max_features,
             "max_depth":max_depth,
             "min_samples_split":min_samples_split,
             "min_samples_leaf":min_sample_leaf
            }
```

```python
# Random search of Parameters, using 5 fold cross validation,
# Search across 100 different combinations
rf_random=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,scoring="neg_mean_squared_error",n_iter=10,cv=5,verbose=2,random_state=42,n_jobs=1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
prediction = rf_random.predict(X_test)
```

```python
recall_score(y_test,rf_random.predict(X_test))
```
